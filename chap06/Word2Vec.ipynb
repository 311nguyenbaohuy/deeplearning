# **Brown corpus**
The Brown Corpus was the first **million-word** electronic corpus of English, created in 1961 at Brown University. This corpus contains text from 500 sources, and the sources have been categorized by genre, such as news, editorial, and so on. 1.1 gives an example of each genre (for a complete list, see http://icame.uib.no/brown/bcm-los.html).



```
nltk.download('brown')
from nltk.corpus import brown
w2v_model = Word2Vec(brown.sents(), size=128, window=5, min_count=3, workers=4)
```

```
# Getting most similar vectors
print(w2v_model.wv.most_similar('Paris'))
```

```
print(w2v_model.wv.most_similar(positive=['woman', 'king'], negative=['man']))
```

# **Text8Corpus**
The file you listed above, has 253,855 unique words. Total number of words is 17,005,208.

```
from gensim.models.word2vec import Text8Corpus
 
# Go here and download + unzip the Text8 Corpus: http://mattmahoney.net/dc/text8.zip
# We take only words that appear more than 150 times for doing a visualization later
w2v_model2 = Word2Vec(Text8Corpus('/content/drive/My Drive/Colab Notebooks/Chap06/text8'), size=100, window=5, min_count=150, workers=4)
```

```
# Getting most similar vectors
print(w2v_model2.wv.most_similar('paris'))

 
```

```

# "King" - "Man" + "Woman" == "Queen"
print(w2v_model2.wv.most_similar(positive=['woman', 'king'], negative=['man']))
```

```
# "Father" - "Boy" + "Girl" == "Mother"
print(w2v_model2.most_similar(['girl', 'father'], ['boy'], topn=3))

# "Paris" - "France" + "Italy" = "Rome"
print(w2v_model2.most_similar(['paris', 'italy'], ['france'], topn=3))

 
```
